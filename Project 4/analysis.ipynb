# Project 4: Employee Attrition Insights

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('IBM_-HR-Employee-Attrition.csv', encoding='iso-8859-1')
print(df.info())
```

## Attrition Rate

```python
df['Attrition'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

## Department Distribution

```python
department_by_attrition = df.groupby(['Department', 'Attrition']).size().unstack()
department_by_attrition.plot(kind='bar', color=['blue', 'orange'])
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

## Total Employees by OverTime

```python
employees_by_attrition = df.groupby('OverTime')['EmployeeCount'].count()
employees_by_attrition.plot(kind='pie', autopct='%1.1f%%', startangle=90, color=['blue', 'orange'])
plt.title('Total Employees by OverTime')
plt.axis('equal')
plt.show()
```
